# Tuple
### Immutable
### Ordered
### Heteregeneous
### Allows Dupliactes

```python
t1=(10,20,30)
print(type(t1))
t2=()                #empty tuple
print(type(t2))
t3=(10)              #tuple having single value
print(type(t3))
t4=(10,)             #correct way to define single value tuple
print(type(t4))
```

```python
t5=(10,20,30,40,50) 
# Slicing for Tuple
print(t5[1])
print(t5[-1])
print(t5[1:3])
```

```python
tup=(10,20,30,30,40,20,50,30,60,70)
print(tup.count(30))      # Gives the count of the element
print(tup.index(60))      # First occurence of the element
print(len(tup))           # Gives the length of the Tuple
print(max(tup))           # Gives the element which has the maximum value
print(min(tup))           # Gives the element which has the minimum value
print(sum(tup))           # Gives the sum of all elements
```

# Dictionary

### key acts as index value, it should be unique

#### empty dictionary

```python
d1={}
print(d1)
print(type(d1))

d2=dict()
print(d2)
print(type(d2))
```

```python
d1['city']='pune'
print(d1)
```

### methods

```python
stud={'name':'alice','age':23,'city':'pune'}
print(stud)
print(stud.keys())                   #only keys
print(stud.values())                 #only values
print(stud.items())                  #key-value as tuple
```

## Methods 
### inserting elements in a set

```python
#add elements
set1={"a","b","c","d","e","f","g"}
print(set1)
set1.add("h")
print(set1)

set1.update("i","j","k")
print(set1)
```

### removing elements from a set

```python
set1.remove("d")          #gives error if element is not present
print(set1)

set1.discard("m")         #no errors 
print(set1)

set1.pop()                #random element will be removed
print(set1)

set1.clear()              #empties the set completely
print(set1)
```

### editing elements of a set
###### editing directly is not allowed
###### we have to remove the element and then add the element again
## Operations on sets

```python
se1={10,20,30,40,50,60}
se2={60,40,30,70,80,90}
```

### union

```python
print(se1.union(se2))
print(se1|se2)
```

### intersection

```python
print(se1.intersection(se2))
print(se1 & se2)
```

### difference

```python
print(se1.difference(se2))
print(se1-se2)

print(se2.difference(se1))
print(se2-se1)
```

### symmetric difference

```python
print(se1.symmetric_difference(se2))
print(se1 ^ se2)
```
