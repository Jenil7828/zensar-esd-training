```python
x = int(input("Enter the 1st number")) # Takes input as string and converts then by type casting into integar
y = int(input("Enter the 2nd number"))
print("""
Basic Calculator
1. Add
2. Subtract
3. Multiply
4. Division
5. Exponent
6. Floor Division
7. Modulus
8. Exit
""") # This is a Multi-line string used to print a menu driven output
choice = int(input("Enter Choice:")) # Takes input for the operation to be performed
# These are Arithmetic Operations
if choice == 1:
    print("Addition :", x+y) # + is addition operator
elif choice == 2:
    print("Subtraction:", x-y) #- is Subtraction operator
elif choice == 3:
    print("Multiplication:", x*y) # * is Multiplication operator
elif choice == 4:
    print("Division:", x/y) # / is Division operator
elif choice == 5:
    print("Exponent:", x**y) # ** is expoenent operator
elif choice == 6:
    print("Floor Divison:", x//y) # // is floor division operator
elif choice == 7:
    print("Modulus", x%y) # is modulus operator
elif choice == 8:
    exit()
else:
    print("Invalid Input")
```

```python
# Understanding nested if-else
no1= int(input("enter a no:"))
flag=1
#check if prime or no
if no1==0:
    print("you entered a 0")
elif no1 == 1:
    print("you have entered 1")
else:
    for i in range(2,no1):
        if no1%i == 0:
            flag = 0
            break
    if flag == 0:
        print("NOT PRIME")
    else:
        print("PRIME")
```
