```python
class Vehicle:
    def __init__(self, name, model):
        self.name = name
        self.model = model
    def display(self):
        print(f'Model Name: {self.name}, Model Number: {self.model}')

v = Vehicle("Toyota", 6349823723)
v.display()
```

```python
# Inheritance of class Vehicle to Car
class Car(Vehicle):
    def __init__(self, name, model, sitting_capacity):
        super().__init__(name, model) # Helps in Inheritance by calling the funtion of the parent class
        self.sitting_capacity = sitting_capacity
    def display(self):
        super().display()
        print(f'Sitting Capacity: {self.sitting_capacity}')
```

```python
c = Car("Toyota", "Etios", 5)
c.display()
```

```python
class Computer:
    def __init__(self):
        # Encapulation of the maximum price from others
        self.__max_price = 900 # double underscope gives private or secure variable
    
    def sell(self):
        print("Selling Price: {}".format(self.__max_price))
    def set_max_price(self, price):
        self.__max_price = price # Changes the price of private variable 
```

```python
co = Computer() #Declare an instaqnce of class
co.sell() # Use the method sell to display the price of computer
co.set_max_price(1000) # # Changes the price of computer
co.sell()  # Use the method sell to display the price of computer
```

```python
import math
class Circle:
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi*self.radius**2
class Rectangle:
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
    def area(self):
        return self.length* self.breadth

# Polymorphism
# Funtion to execute polymorphism
def area_shape(shape):
    #Funtion to compute area of any shape by using polymorphiosm
    print(f'Area of: {shape.area()}')

cir = Circle(20)
rec = Rectangle(10, 20)
area_shape(cir) # Uses Polymorphism by assigning shape 
area_shape(rec)
```

```python
class A:
    def greet(self):
        print("Hi")

class B(A):
    def greet(self):
        print("Hello")
    def greet1(self):
        print("GoodBye")

# Polymorphism as the same method greet is A as well as B which results in Overriding
g = B()
g.greet()
g.greet1()
g1 = A()
g1.greet()
```

# Exception Handling

```python
# try-except block
try:
    my_list = [1, 2, 3, 4]
    print(my_list[4])
except IndexError as e:
    print(f"Error- {e}")
```

```python
# Single try Mulitple except block 
try:
    my_list = [1, 2, 3, 4]
    print(my_list[4])
    t = 200/0
    d = {"Jenil": "Rathod"}
    print(d["Zaid"])
except IndexError as e1:
    print(f"Error- {e1}")
except KeyError as e2:
    print(f"Error- {e2}")
except ZeroDivisionError as e3:
    print(f"Error- {e3}")
```

```python
# Single try-except block with multiple condition in single except 
try:
    my_list = [1, 2, 3, 4]
    t = 200/0
    print(my_list[4])
    d = {"Jenil": "Rathod"}
    print(d["Zaid"])
except (IndexError, KeyError, ZeroDivisionError) as e:
    print(f"Error- {e}")
```

```python
# try-except-finally block
try:
    my_list = [1, 2, 3, 4]
    print(my_list[4])
except IndexError as e:
    print(f"Error- {e}")
finally:
    print("Program finished")
```

```python
# try-except-else-finally block
try:
    my_list = [1, 2, 3, 4]
    print(my_list[3])
except IndexError as e:
    print(f"Error- {e}")
else:
    print("No Error Found")
finally:
    print("Program finished")
```

# File Handling

## Process to Hanle files
### 1. Open the file
### 2. Do your work
### 3. Close your file

```python
file = open("example.txt", 'w') # open funtion is used to open a file from the system and r, w, r+, w+, a, a+, etc., are the different modes 
txt = "Write this into the file"
file.write(txt) # Writes the text into the file
file.close() # Close the file
```

# Mode  	Description
### " r "       Read mode. Opens a file for reading. The file must exist.
### " w "	    Write mode. Opens a file for writing. Creates a new file if it doesn’t exist or truncates the file if it exists.
### " a "	    Append mode. Opens a file for appending. Creates a new file if it doesn’t exist.
### " b	"     Binary mode. Opens a file in binary mode.
### " t "	    Text mode. Opens a file in text mode. Default mode.
### " x "	    Exclusive creation mode. Creates a new file and fails if it already exists.
### " r+ "	    Read and write mode. Opens a file for both reading and writing. The file must exist.
### " w+ "	    Write and read mode. Opens a file for both writing and reading. Creates a new file if it doesn’t exist or truncates the file if it exists.
### " a+ "	    Append and read mode. Opens a file for appending and reading. Creates a new file if it doesn’t exist
