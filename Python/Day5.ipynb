### Opening file using With

```python
# file is automatically closed
```

```python
with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

### Exception handling in files

```python
try:
    with open('nonexistent.txt','r') as file:
        content=file.read()
except FileNotFoundError:
    print('file does not exists')
except IOError:
    print('error occured while handling the file')
else:
    print(content)
finally:
    print("operation completed")
```

### File positioning

```python
# seek(offset,from where):move pointer
# tell:returns current position of pointer
```

```python
with open('example.txt','r') as file:
    print(file.read(5))
    print(file.tell())           #current position of pointer
    file.seek(0)                 #move pointer to start
    print(file.read(5))
    file.seek(5,0)               #move pointer to 5th index from current index
    print(file.read(5))
```

### Deleting files

```python
# using os module
```

```python
import os
if os.path.exists('example.txt'):
    os.remove('example.txt')
    print('file deleted')
else:
    print('file does not exist')
```

## Handling Missing Values

```python
data = {
    'Name': ['Jenil', 'Devashree', 'Zaid'],
    'Age': [np.nan, 20, 20],
    'Salary': [50000, 20000, 1000000],
    'Department' : ['ML', 'FullStack', np.nan]
}
df = pd.DataFrame(data)
print('Original DataFrame')
print(df)
 # Filling NaN values with a specific values
print("\n\nData Frame after filling missing values by 0\n",df.fillna(0))

# Dropping Rows with NaN values 
print("\n\n",df.dropna())
# Dropping columns with NaN values
print("\n\n",df.dropna(axis=1))
```

## Merging and Joining Data Frames
### Merging Data Frames

```python
data1 = {'id': [1, 2, 3, 4],
         'name': ['Jenil' , 'Kavan', 'Lara', 'Mangaya']
        }
df1 = pd.DataFrame(data1)

data2 = {'id': [1, 2, 3, 4, 5],
         'Salary': [50000, 40000, 30000, 20000, 10000]
        }
df2 = pd.DataFrame(data2)
merge = pd.merge(df1, df2, on='id', how='inner')
print('Merged Data Set is:\n')
print(merge)
merge = pd.merge(df1, df2, on='id', how='outer')
print('\n Merged Data Set is:\n')
print(merge)
```

## Visualization with Pandas

```python
import matplotlib.pyplot as plt
```

### Line Plot

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
# Ploting the Data
plt.plot(x, y, label="Line Plot", color = 'red', marker = 'o')

#Adding Labels and Titles
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Basic Line Plot")

# Adding a Legend
plt.legend()

#Display the plot
plt.show()
```

## Customizing plots
### Adding grids, Line Style and Markers with multiple lines

```python
# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
y1 = [1, 3, 5, 7, 9]
# Ploting the Data
plt.plot(x, y, label="Line Plot", linestyle = "--", color = 'red', marker = 'o')
plt.plot(x, y1, label="Line 2", linestyle = '-', color = 'pink', marker = 'x')
#Adding Labels and Titles and Grid
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Customized Line Plot")
plt.grid(True)
# Adding a Legend
plt.legend()

#Display the plot
plt.show()
```

## Bar Charts

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 5, 7, 9]

# Vertical Bar Chart
plt.bar(categories, values, color = 'greenyellow')
plt.title("Vertical Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()

# Horizontal Bar Chart
plt.barh(categories, values, color = 'yellow')
plt.title("Horizontal Bar Chart")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## Creating a 3D Plot
