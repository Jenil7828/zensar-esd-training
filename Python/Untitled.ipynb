```python
def vowel_count():
    string = input("Enter the string you want to seperate vowels from: ")
    vowel_c = 0
    consonent_c = 0
    for i in string:
        if i in "AEIOUaeiou":
            vowel_c+=1
        elif i ==" ":
            continue
        else:
            consonent_c+=1
    return f"The number of vowels are: {vowel_c}\nThe number of consonents are: {consonent_c}"
    
```

```python
print(vowel_count())
```

```python
def vowel_sep():
    string = input("Enter the string you want to seperate vowels from: ")
    vowel_c = ""
    consonent_c = ""
    for i in string:
        if i in "AEIOUaeiou":
            vowel_c+=i
        else:
            consonent_c+=i
    return f"The vowels apperaing are: {vowel_c}\nThe appearing consonents are: {consonent_c}"

print(vowel_sep())
```
